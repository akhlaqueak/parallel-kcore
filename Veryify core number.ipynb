```python
import networkx as nx
import os
```

```python
dataset = "roadNetPa.g"
```

```python
G = nx.read_adjlist("data_set/data/ours_format/" + dataset)
```

```python
nxkcore = nx.core_number(G)
```

```python
d = {}
with open("k-core.txt") as f:
    d[f.readline().strip()] = 0
    for line in f:
       (key, val) = line.split()
       d[key] = int(val)
if d == nxkcore:
    print("Test Passed!")
    print(nxkcore)
else:
    print("Test Failed!")
    print(set(d.items()) - set(nxkcore.items()))
```
